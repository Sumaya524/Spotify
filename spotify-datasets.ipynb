```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks=pd.read_csv("/kaggle/input/spotify-tracks-datasets/tracks.csv")
```

```python
df_tracks.head()
```

```python
# checking for sum of null values

pd.isnull(df_tracks).sum()
```

## So there are 71 missing values in the "name" column.

```python
df_tracks.info()
```

## let's find the least popular songs from the column "popularity"

```python
sorted_df=df_tracks.sort_values("popularity", ascending=True).head(10)
sorted_df
```

### Let's find the most 10 popular songs which are >90 in descending order.
### Here we are using query() function


```python
most_popular=df_tracks.query('popularity>90', inplace=False).sort_values('popularity',ascending=False)
most_popular.head(10)
```

## move the  the release_date to first column
## The set_index method allows  one or more column values become the row index

```python
df_tracks.set_index('release_date', inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

## Suppose we want to check the location of specfic variable, use the 'iloc()' method

## suppose we want to check the name of the artist in the 18th row

```python
df_tracks[["artists"]].iloc[18]
```

## Next we are converting 'duration_ms' milliseconds to seconds, use lambda() function

```python
df_tracks['duration']=df_tracks[["duration_ms"]].apply(lambda x:round(x/1000))
df_tracks.drop(["duration_ms"],axis=1,inplace=True)
```

```python
df_tracks['duration'].head()
```

# Visualization
## Correlation map
### We are dropping three columns "key", "explicit", "mode"
### Pearson Correlation method
### heatmap

```python
corr_df=df_tracks.drop(["key", "mode", "explicit"],axis=1).corr(method="pearson")

plt.figure(figsize=(14,6))

heatmap=sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno",
                   linewidths=1, linecolor="Black")

heatmap.set_title("correlation heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

## Let's create a sample dataset from the original dataset of 4%

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

# Regression plot
## Now we are creating a regression plot of "loudness" and "energy", beacuse there is a positive correlation between them

```python
sns.regplot(x="loudness", y="energy", data=sample_df).set(title="Loudness Vs Energy")
plt.figure(figsize=(10,6))
```

### There is a positive correlation between them

# Next we see the correlation between "acousticness" and "popularity"

```python
sns.regplot(x="acousticness", y="popularity", data=sample_df,color="b").set(title="popularity Vs acousticness correlation")
plt.figure(figsize=(10,6))
```

### here there is negative correlation between them

## Next we are creating a new column "year" from "release_date" column¶

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
df_tracks['dates'].head()
```

# Distribution Plot

## Number of songs per year

```python
sns.displot(years,discrete=True,aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

# Barplot


## We will create a barplot between "duration of songs and Year"
## we changed the "duration_ms" to "duration" column

```python
total_dr=df_tracks.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years, y=total_dr,ax=ax, errwidth=False).set(title="Years Vs Duration")
plt.xticks(rotation=90)
```

## We will see the SpotifyFeatures dataset

```python
df_genre=pd.read_csv("/kaggle/input/spotify-features/SpotifyFeatures.csv")
```

```python
df_genre.head()
```

# Barplot

## Let's find Duration of the songs for different Genre

```python
sns.barplot(data=df_genre, x="duration_ms", y="genre")
sns.color_palette("rocket", as_cmap=True)
plt.title("Duration of the songs in Different Genres")
plt.xlabel("duration in milli seconds")
plt.ylabel("Genres")
```

## let's find the Top 5 popularity songs based on genre

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous=df_genre.sort_values("popularity", ascending=False).head(10)
sns.barplot(x="popularity", y="genre", data=famous).set(title="Top 5 Genres based on Popularity")
```
